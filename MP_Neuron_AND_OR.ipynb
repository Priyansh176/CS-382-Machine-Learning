# McCulloch-Pitts (MP) Neuron Model

Simple implementation for AND / OR logic using thresholding.

```

def mp_neuron(x1, x2, weights, threshold):
    # Weighted sum
    net = x1 * weights[0] + x2 * weights[1]
    
    # Activation (step function)
    if net >= threshold:
        return 1
    else:
        return 0
```

## AND Gate

```

# AND gate parameters
weights_and = [1, 1]
threshold_and = 2

# Test cases
inputs = [(0,0), (0,1), (1,0), (1,1)]

print("AND Gate Results:")
for x1, x2 in inputs:
    print(x1, x2, "->", mp_neuron(x1, x2, weights_and, threshold_and))
```

## OR Gate

```

# OR gate parameters
weights_or = [1, 1]
threshold_or = 1

print("OR Gate Results:")
for x1, x2 in inputs:
    print(x1, x2, "->", mp_neuron(x1, x2, weights_or, threshold_or))
```
